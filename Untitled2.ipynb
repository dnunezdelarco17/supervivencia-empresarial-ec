```python
# 1) Sube el archivo BIZ_CASEPSSPerfAnalysis.xlsx
from google.colab import files
uploaded = files.upload()          # selecciona tu .xlsx
ARCHIVO = list(uploaded.keys())[0]
print("Archivo cargado:", ARCHIVO)
```

```python
# 2) Carga, limpieza y bandera de mes parcial
import pandas as pd, numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel(ARCHIVO, sheet_name='Data', engine='openpyxl')
df = df.loc[:, ~df.columns.astype(str).str.contains('^Unnamed')]
df.columns = [str(c).strip() for c in df.columns]
df = df.rename(columns={'meal_bookings (USD)':'gmv', 'Ads USD':'ads_usd',
                        'Restaurant_cancellations':'restaurant_cancellations'})

num = ['requested','completed_trips','total_users','gmv','menu_available_minutes',
       'restaurant_visible_minutes','rating','store_impressions','order_accuracy',
       'ads_usd','restaurant_cancellations']
for c in num:
    df[c] = pd.to_numeric(df[c], errors='coerce')

df['month_date'] = pd.to_datetime(df['month_date'])
df = df.dropna(subset=['store_name']).reset_index(drop=True)

# Bandera mes parcial: GMV del store-mes < 40% de la mediana del store
med = df.groupby('store_name')['gmv'].transform('median')
df['es_mes_parcial'] = df['gmv'] < 0.4 * med

print(f"Registros: {len(df)} | Tiendas: {df['store_name'].nunique()} | "
      f"Periodo: {df['month_date'].min().date()} -> {df['month_date'].max().date()}")
print("Meses marcados como PARCIALES (excluidos del crecimiento):")
print(df.loc[df.es_mes_parcial, ['month_date','store_name','gmv']].to_string(index=False))
```

```python
# 3) KPIs a nivel fila (para detalle, export y gráfico de evolución)
df['completion_rate']   = df['completed_trips'] / df['requested'] * 100
df['availability_rate'] = df['restaurant_visible_minutes'] / df['menu_available_minutes'] * 100
df['conversion_rate']   = df['completed_trips'] / df['store_impressions'] * 100   # recalculado
df['order_accuracy_pct']= (df['completed_trips'] - df['order_accuracy']) / df['completed_trips'] * 100
df['cancellation_rate'] = df['restaurant_cancellations'] / df['requested'] * 100
df['aov']               = df['gmv'] / df['completed_trips']
df['revenue_per_user']  = df['gmv'] / df['total_users']
df[['store_name','month_date','completion_rate','availability_rate','conversion_rate',
    'order_accuracy_pct','rating','cancellation_rate','aov']].head()
```

```python
# 4) Resumen PONDERADO POR VOLUMEN (cociente de sumas) -> reproduce las cifras del deck
agg = df.groupby('store_name').agg(
    GMV_USD   = ('gmv','sum'),
    req       = ('requested','sum'),
    completed = ('completed_trips','sum'),
    impr      = ('store_impressions','sum'),
    avail_min = ('menu_available_minutes','sum'),
    vis_min   = ('restaurant_visible_minutes','sum'),
    canc      = ('restaurant_cancellations','sum'),
    probs     = ('order_accuracy','sum'),
    users     = ('total_users','sum'),
    ads       = ('ads_usd','sum'),
)
# rating ponderado por nº de órdenes
rating_w = (df.groupby('store_name')[['rating','completed_trips']]
              .apply(lambda g: np.average(g['rating'], weights=g['completed_trips'])))

resumen = pd.DataFrame({
    'GMV_USD':         agg['GMV_USD'].round(0),
    'Completion_%':    (agg.completed/agg.req*100),
    'Availability_%':  (agg.vis_min/agg.avail_min*100),
    'Conversion_%':    (agg.completed/agg.impr*100),
    'OrderAccuracy_%': ((agg.completed-agg.probs)/agg.completed*100),
    'Rating':          rating_w,
    'Cancellation_%':  (agg.canc/agg.req*100),
    'AOV_USD':         (agg.GMV_USD/agg.completed),
    'RevPerUser_USD':  (agg.GMV_USD/agg.users),
}).round(2)
print("=== RESUMEN PONDERADO POR VOLUMEN ===")
print(f"GMV total portafolio: ${df.gmv.sum():,.0f} | "
      f"Conversión global {df.completed_trips.sum()/df.store_impressions.sum()*100:.2f}% | "
      f"AOV global ${df.gmv.sum()/df.completed_trips.sum():.2f}")
resumen
```

```python
# 5) GMV por tienda, total y tendencia mensual del portafolio
gmv_store = df.groupby('store_name')['gmv'].sum().sort_values(ascending=False)
print("GMV por tienda:")
for s,v in gmv_store.items(): print(f"  {s}: ${v:,.0f}")
print(f"  TOTAL: ${gmv_store.sum():,.0f}")

gmv_mes = df.groupby('month_date')['gmv'].sum()
top = gmv_mes[~df.groupby('month_date')['es_mes_parcial'].any()].sort_values(ascending=False).head(3)
print("\nMeses pico (excluye parciales):")
for d,v in top.items(): print(f"  {d.strftime('%b-%y')}: ${v:,.0f}")
```

```python
# 6) Crecimiento a MES COMPLETO (jul-21 -> jun-22). Julio-22 se excluye por parcial.
ini, fin = '2021-07', '2022-06'
def gmv_de(store, ym):
    return df[(df.store_name==store) & (df.month_date.dt.strftime('%Y-%m')==ym)]['gmv'].sum()

print(f"Crecimiento {ini} -> {fin}:")
for s in sorted(df.store_name.unique()):
    f, l = gmv_de(s, ini), gmv_de(s, fin)
    print(f"  {s}: ${f:,.0f} -> ${l:,.0f} = {(l/f-1)*100:+.0f}%")
pf = df[df.month_date.dt.strftime('%Y-%m')==ini]['gmv'].sum()
pl = df[df.month_date.dt.strftime('%Y-%m')==fin]['gmv'].sum()
print(f"  PORTAFOLIO: ${pf:,.0f} -> ${pl:,.0f} = {(pl/pf-1)*100:+.0f}%")
```

```python
# 7) Ingreso RECUPERABLE llevando disponibilidad a 98% (cero inversión de marketing)
META_DISP = 0.98
filas, total = [], 0.0
for s in sorted(df.store_name.unique()):
    g = df[df.store_name==s]
    disp = g.restaurant_visible_minutes.sum() / g.menu_available_minutes.sum()
    gmv  = g.gmv.sum()
    rec  = (META_DISP - disp) * gmv / disp        # GMV proporcional a minutos visibles extra
    total += rec
    filas.append((s, f"{disp*100:.1f}%", f"${rec:,.0f}"))
rec_df = pd.DataFrame(filas, columns=['Tienda','Disponibilidad','Recuperable_anual'])
print(rec_df.to_string(index=False))
print(f"\nTOTAL recuperable a 98%: ${total:,.0f}  (sin spend de marketing)")
```

```python
# 8) Eficiencia de Ads — UN solo método: proxy total/total (NO es ROAS atribuido)
ads = df.groupby('store_name').agg(ads=('ads_usd','sum'), gmv=('gmv','sum'))
ads['GMV_por_$1_ads'] = (ads.gmv/ads.ads).round(1)
ads['ads_cost_pct']   = (ads.ads/ads.gmv*100).round(2)
print("Proxy de eficiencia publicitaria (cota superior, no atribuido):")
print(ads.round(1).to_string())
print(f"\nBlended portafolio: {df.gmv.sum()/df.ads_usd.sum():.1f}x  | "
      f"Ads = {df.ads_usd.sum()/df.gmv.sum()*100:.2f}% del GMV")
```

```python
# 9) Scorecard normalizado a meta, pesos suman 1.0
pesos = {
    'Completion_%':    0.20,
    'Availability_%':  0.20,
    'Rating':          0.20,
    'Conversion_%':    0.15,
    'OrderAccuracy_%': 0.15,
    'Cancellation_%':  0.10,
}
assert abs(sum(pesos.values()) - 1.0) < 1e-9, "Los pesos deben sumar 1.0"

metas   = {'Completion_%':98,'Availability_%':98,'Rating':4.8,
           'Conversion_%':8,'OrderAccuracy_%':99,'Cancellation_%':1}
invert  = {'Cancellation_%'}   # menor es mejor

sc = pd.DataFrame(index=resumen.index)
for k in pesos:
    if k in invert:
        sc[k] = (metas[k] / resumen[k]).clip(upper=1) * 100
    else:
        sc[k] = (resumen[k] / metas[k]).clip(upper=1) * 100
sc['SCORE_/100'] = sum(sc[k]*w for k,w in pesos.items())
sc = sc.round(1).sort_values('SCORE_/100', ascending=False)
print("Score global por tienda (0-100):")
sc[['SCORE_/100']]
```

```python
# 10) Visualizaciones
# (a) Tendencia de GMV por tienda, marcando el mes parcial
piv = df.pivot_table(index='month_date', columns='store_name', values='gmv', aggfunc='sum')
ax = piv.plot(figsize=(13,5), marker='o')
parc = df.loc[df.es_mes_parcial,'month_date'].min()
if pd.notna(parc):
    ax.axvspan(parc, piv.index.max(), color='grey', alpha=0.15)
    ax.text(parc, ax.get_ylim()[1]*0.92, '  mes parcial (excluido del crecimiento)', fontsize=9)
ax.set_title('GMV mensual por tienda', fontweight='bold'); ax.set_ylabel('USD'); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()

# (b) KPIs por tienda (ponderados)
kpis = ['Completion_%','Availability_%','Conversion_%','OrderAccuracy_%','Rating','Cancellation_%']
fig, axes = plt.subplots(2,3, figsize=(15,7))
for axx,k in zip(axes.flat, kpis):
    resumen[k].plot(kind='bar', ax=axx, color='#06C167', edgecolor='black')
    axx.set_title(k, fontweight='bold'); axx.set_xlabel(''); axx.tick_params(axis='x', rotation=0)
plt.tight_layout(); plt.show()
```

```python
# 11) Exportar resultados
with pd.ExcelWriter('Resultados_KPIs.xlsx') as w:
    resumen.to_excel(w, sheet_name='Resumen_ponderado')
    sc.to_excel(w, sheet_name='Scorecard')
    rec_df.to_excel(w, sheet_name='Recuperable_98', index=False)
    df.to_excel(w, sheet_name='Detalle_calculado', index=False)
from google.colab import files
files.download('Resultados_KPIs.xlsx')
print('Listo: Resultados_KPIs.xlsx')
```

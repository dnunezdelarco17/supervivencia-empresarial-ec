```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel(ARCHIVO, sheet_name='Data', engine='openpyxl')

# Quitar columnas vacías y normalizar nombres
df = df.loc[:, ~df.columns.astype(str).str.contains('^Unnamed')]
df.columns = [str(c).strip() for c in df.columns]
df = df.rename(columns={'meal_bookings (USD)': 'meal_bookings', 'Ads USD': 'ads_usd'})

# Tipos numéricos (conversion_rate se recalcula, no se usa la del archivo)
num_cols = ['requested','completed_trips','total_users','meal_bookings',
            'menu_available_minutes','restaurant_visible_minutes','rating',
            'store_impressions','order_accuracy','ads_usd','Restaurant_cancellations']
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors='coerce')

df['month_date'] = pd.to_datetime(df['month_date'])
df = df.dropna(subset=['store_name']).reset_index(drop=True)

print(f"Registros: {len(df)}  |  Tiendas: {df['store_name'].nunique()}")
print(f"Periodo: {df['month_date'].min().date()} -> {df['month_date'].max().date()}")
df.head()
```

```python
# 1. Completion Rate = completed_trips / requested * 100
df['completion_rate'] = df['completed_trips'] / df['requested'] * 100

# 2/3. Disponibilidad real = minutos reales / minutos programados * 100
df['availability_rate'] = df['restaurant_visible_minutes'] / df['menu_available_minutes'] * 100

# 4. AOV (Average Order Value) = ventas / pedidos completados
df['aov'] = df['meal_bookings'] / df['completed_trips']

# 5. Conversion Rate = completed_trips / store_impressions * 100
df['conversion_rate'] = df['completed_trips'] / df['store_impressions'] * 100

# 6. Order Accuracy = (completados - pedidos con problema) / completados * 100
df['order_accuracy_pct'] = (df['completed_trips'] - df['order_accuracy']) / df['completed_trips'] * 100

# 7. Customer Rating (ya es promedio)
df['customer_rating'] = df['rating']

# 8. Restaurant Cancellation Rate = cancelaciones / requested * 100
df['cancellation_rate'] = df['Restaurant_cancellations'] / df['requested'] * 100

# 9. Revenue per User = ventas / usuarios
df['revenue_per_user'] = df['meal_bookings'] / df['total_users']

# 10. Ads: ROAS real NO calculable -> proxy y Ads Cost %
df['ads_cost_pct'] = df['ads_usd'] / df['meal_bookings'] * 100
df['roas_proxy']   = df['meal_bookings'] / df['ads_usd']   # cota superior, NO es ROAS real

df[['store_name','month_date','completion_rate','availability_rate',
    'conversion_rate','order_accuracy_pct','customer_rating',
    'cancellation_rate','aov','revenue_per_user']].head()
```

```python
# Metas (estándar de excelencia del documento)
metas = {
    'Completion_Rate':   ('>=', 98),
    'Availability':      ('>=', 98),
    'Conversion_Rate':   ('>=', 8),
    'Order_Accuracy':    ('>=', 99),
    'Rating':            ('>=', 4.8),
    'Cancellation_Rate': ('<=', 1),
    'ROAS_proxy':        ('>=', 5),
}

def estado(valor, regla):
    op, meta = regla
    cumple = valor >= meta if op == '>=' else valor <= meta
    return ('✅' if cumple else '❌') + f' (meta {op} {meta})'

score = resumen[list(metas.keys())].copy()
for k, regla in metas.items():
    score[k] = score[k].map(lambda v: f'{v:.2f}  ' + estado(v, regla))
score.T
```

```python
# Pesos del scorecard recomendado
pesos = {
    'completion_rate':    0.20,
    'availability_rate':  0.15,   # se usa como proxy de Menu Availability + Visibility
    'customer_rating':    0.20,
    'order_accuracy_pct': 0.10,
    'conversion_rate':    0.10,
    'cancellation_rate':  0.05,
    'roas_proxy':         0.05,
}
# Metas para normalizar cada KPI a 0-100 (cap a 100)
metas_norm = {
    'completion_rate':98,'availability_rate':98,'customer_rating':4.8,
    'order_accuracy_pct':99,'conversion_rate':8,'cancellation_rate':1,'roas_proxy':5,
}
invertidos = {'cancellation_rate'}  # menor es mejor

p = resumen.index.to_frame(index=False)
base = df.groupby('store_name').mean(numeric_only=True)

puntajes = {}
for kpi, w in pesos.items():
    meta = metas_norm[kpi]
    val = base[kpi]
    pct = (meta / val).clip(upper=1) if kpi in invertidos else (val / meta).clip(upper=1)
    puntajes[kpi] = pct * 100

tabla_p = pd.DataFrame(puntajes)
tabla_p['SCORE_PONDERADO'] = sum(tabla_p[k]*w for k,w in pesos.items())
tabla_p = tabla_p.round(1).sort_values('SCORE_PONDERADO', ascending=False)
print("Puntaje global por tienda (0-100):")
tabla_p[['SCORE_PONDERADO']]
```

```python
kpis_plot = ['completion_rate','availability_rate','conversion_rate',
             'order_accuracy_pct','customer_rating','cancellation_rate']
titulos = ['Completion Rate (%)','Availability (%)','Conversion Rate (%)',
           'Order Accuracy (%)','Customer Rating','Cancellation Rate (%)']

fig, axes = plt.subplots(2, 3, figsize=(15, 7))
for ax, kpi, tit in zip(axes.flat, kpis_plot, titulos):
    base[kpi].plot(kind='bar', ax=ax, color='#3b6fb5', edgecolor='black')
    ax.set_title(tit, fontweight='bold'); ax.set_xlabel('')
    ax.tick_params(axis='x', rotation=0)
plt.tight_layout(); plt.show()
```

```python
# Evolución mensual del Completion Rate por tienda
piv = df.pivot_table(index='month_date', columns='store_name',
                     values='completion_rate', aggfunc='mean')
piv.plot(figsize=(13,5), marker='o')
plt.axhline(98, ls='--', color='red', label='Meta 98%')
plt.title('Completion Rate mensual por tienda', fontweight='bold')
plt.ylabel('%'); plt.legend(); plt.grid(alpha=0.3); plt.show()
```
